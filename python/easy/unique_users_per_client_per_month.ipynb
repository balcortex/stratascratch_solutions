```python
import pandas as pd

# Read data
fact_events = pd.read_csv("./unique_users_per_client_per_month.csv")
fact_events["time_id"] = pd.to_datetime(fact_events.time_id)
fact_events.head()
```

```python
# Add column month to group by it later
fact_events["month"] = fact_events.time_id.dt.month
fact_events.head()
```

```python
# Group by client_id and month, then count the unique elements
result = fact_events.groupby(["client_id", "month"]).nunique()

# Drop the columns we are not interested in
result = result.drop(columns=["id", "customer_id", "event_type", "event_id", "time_id"])

# Reintegrate the indices (groups) into the columns
result = result.reset_index()

# Rename the column month to match the expected output name
result = result.rename(columns={"month": "time_id"})

result.head()
```
